```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from math import pi, cos, sin

#   FFT
def reversebits(x):
    log2 = int(np.log2(len(x)))
    for i in range(len(x)):
        k = int((bin(i)[2:]).zfill(log2)[::-1], 2)
        if k > i:
            x[i], x[k] = x[k], x[i]
    return x
        
def unrecursive_fft(x):
    x = reversebits(x)
    n = len(x)
    log2 = int(np.log2(n))
    for i in range(1, log2 + 1):
        n = 2 ** i
        w = [np.exp(-2j*pi*k/n) for k in range(n // 2)]
        for j in range(0, len(x), n):
            for k in range(n // 2):
                l = w[k] * x[k + j + n // 2]
                x[k + j + n // 2] = x[k + j] - l
                x[k+j] += l
    return x

# 
def my_fft2(image):
    n, m = image.shape
    size_n = 2**int(np.ceil(np.log2(n))) #     
    size_m = 2**int(np.ceil(np.log2(m)))
    padded = np.zeros((size_n, size_m), dtype=complex) #     ( )
    padded[:n,:m] = image
    
    # FFT  
    temp = np.array([unrecursive_fft(row) for row in padded])
    
    # FFT  
    return np.array([unrecursive_fft(col) for col in temp.T]).T

#  
def fftshift(fft_img):
    n, m = fft_img.shape
    return np.roll(fft_img, (n//2, m//2), axis=(0,1))
# 
def ifft2(fft_img):
    return np.conj(my_fft2(np.conj(fft_img)))/(fft_img.shape[0]*fft_img.shape[1])


image = skimage.io.imread('./cotique.jpeg')
print(image.shape)
image = rgb2gray(image)  #         ()
n, m = image.shape

#  FFT    
fft_image = my_fft2(image.astype(complex))
fshift = fftshift(fft_image)

#  
mask_percent = 0.9
crow, ccol = n//2, m//2
mask = np.zeros_like(fshift)
r, c = int(n*mask_percent/2), int(m*mask_percent/2)
mask[crow-r:crow+r, ccol-c:ccol+c] = 1 #  0  1   

filtered = fshift * mask
reconstructed = np.abs(ifft2(fftshift(filtered)))

# 
plt.figure(figsize=(12,6))
plt.subplot(121), plt.imshow(image, cmap='gray')
plt.subplot(122), plt.imshow(reconstructed, cmap='gray')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft2, ifft2, fftshift
import skimage
from skimage import data
from skimage.color import rgb2gray


image = skimage.io.imread('./cotique.jpeg')
image = rgb2gray(image) #data.astronaut()
plt.figure(figsize=(6,6))
plt.imshow(image, cmap='gray')
plt.title(' ')
plt.axis('off')
plt.show()

#  FFT
fft_image = fft2(image)
fshift = fftshift(fft_image)  #     

#  
magnitude_spectrum = 20 * np.log(np.abs(fshift))
plt.figure(figsize=(12,6))
plt.subplot(121), plt.imshow(magnitude_spectrum, cmap='gray')
plt.title(' '), plt.axis('off')

#   
rows, cols = image.shape
crow, ccol = rows//2, cols//2
mask_percent = 0.5  #  50%  
mask = np.zeros((rows, cols), np.uint8)
r = int(rows * mask_percent / 2)
c = int(cols * mask_percent / 2)
mask[crow-r:crow+r, ccol-c:ccol+c] = 1

filtered_fshift = fshift * mask
plt.subplot(122), plt.imshow(20*np.log(np.abs(filtered_fshift)), cmap='gray')
plt.title(' '), plt.axis('off')
plt.show()

#     
filtered_fft = fftshift(filtered_fshift)
reconstructed_image = np.abs(ifft2(filtered_fft))

plt.figure(figsize=(12,6))
plt.subplot(121), plt.imshow(image, cmap='gray')
plt.title(' '), plt.axis('off')
plt.subplot(122), plt.imshow(reconstructed_image, cmap='gray')
plt.title(f'  ({mask_percent*100:.0f}% )')
plt.axis('off')
plt.show()

# %% [markdown]
# ### 5.   
mse = np.mean((image - reconstructed_image) ** 2)
compression_ratio = (1 - mask_percent) * 100

print(f'MSE: {mse:.20f}')
print(f': {compression_ratio:.0f}%  ')

# %% [markdown]
# ## 
# - FFT      
# -       
# -       
# -        
```

```python
from scipy.fft import ifft, fft, fftfreq
import numpy as np
import matplotlib.pyplot as plt

#     
n = 1000
freqs = np.zeros(n)
freqs[50:55] = 1  # 50-55 Hz
freqs[200:205] = 0.5  # 200-205 Hz
#  
signal = np.real(ifft(freqs))

# 
plt.figure(figsize=(12,4))
plt.plot(signal)
plt.title(' ')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
#    
t = np.arange(365*4)  # 4   
signal = (np.sin(2*np.pi*t/365) +  #  
          0.5*np.sin(2*np.pi*t/7) +  #  
          np.random.normal(0, 0.3, len(t)))

# FFT 
fft_vals = fft(signal)
freqs = fftfreq(len(t))

#   
plt.figure(figsize=(12,8))
plt.subplot(211)
plt.title('')
plt.plot(signal)
plt.subplot(212)
plt.plot(1/freqs[1:len(t)//2], np.abs(fft_vals)[1:len(t)//2])
plt.xscale('log')
plt.title('  ')
plt.xlabel(' ()')
plt.ylabel('')
plt.axvline(365, c='r', ls='--', label='')
plt.axvline(7, c='g', ls='--', label='')
plt.legend()
plt.show()
```

```python
#   
t = np.linspace(0, 1, 1000)
clean = 0.5*np.sin(2*np.pi*10*t) + 0.2*np.sin(2*np.pi*50*t)
noisy = clean + 0.3*np.random.randn(len(t))

#   FFT
fft_signal = fft(noisy)
freqs = fftfreq(len(t), t[1]-t[0])
cutoff = 70  #    70 Hz
fft_signal[np.abs(freqs) > cutoff] = 0
filtered = np.real(ifft(fft_signal))

# 
plt.figure(figsize=(12,4))
plt.plot(t, noisy, alpha=0.5, label='')
plt.plot(t, filtered, 'r', linewidth=2, label=' ')
plt.legend()
plt.show()
```

```python
from scipy.io import wavfile
sample_rate, data = wavfile.read('audio.wav')
length = data.shape[0] / sample_rate
print(length)
time = np.linspace(0., length, data.shape[0])
plt.plot(time, data)
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```

```python
freq = fftfreq(len(time), 1/sample_rate)
fft_data = fft(data)
plt.figure(figsize=(12,4))
# plt.plot(k, filtered, 'r', linewidth=2)
plt.plot(freq, abs(fft_data))
plt.title('  ')
plt.show()
```

```python
fft_data[np.abs(freq) < 200] = 0
plt.figure(figsize=(12,4))
plt.plot(freq, abs(fft_data))
plt.title('   ')
plt.show()
```

```python
filtered = np.real(ifft(fft_data))
filtered = np.int16(filtered)
plt.figure(figsize=(12,4))
plt.plot(time, filtered, 'r', linewidth=2)
plt.title(' ')
plt.show()
wavfile.write('filtered_audio.wav', sample_rate, filtered)
```
